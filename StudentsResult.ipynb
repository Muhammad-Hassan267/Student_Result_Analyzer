```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
def menu():
    print('\n---EXAM RESULT ANALYZER---\n')
    print('1. Add new result')
    print('2. View full data set')
    print('3. View subject wise average')
    print('4. View top students in a subject')
    print('5. View failing students')
    print('6. plot marks by subject')
    print('7. Exit')
def add_result():
    name = input('Enter student name: ')
    class_name = input('Enter class name: ')
    subject = input('Enter subject name: ')
    marks = int(input('Enter marks: '))
    new_data = pd.DataFrame({'name': [name], 'class': [class_name], 'subject': [subject], 'marks': [marks]})
    if os.path.exists('studentdata.csv'):
        df = new_data.to_csv('studentdata.csv', mode='a', header=False, index=False)
    else:
        new_data.to_csv('studentdata.csv', index=False)
    print(f'Result for {name} added successfully!')
def view_data():
    if os.path.exists('studentdata.csv'):
        df = pd.read_csv('studentdata.csv')
        print(df)
    else:
        print('No data available.')
def view_subect_average():
    if os.path.exists('studentdata.csv'):
        df = pd.read_csv('studentdata.csv')
        subject = input('Enter subject name: ')
        if subject in df['subject'].values:
            avg_marks = df[df['subject'] == subject]['marks'].mean()
            print(f'Average marks in {subject}: {avg_marks}')
        else:
            print(f'Subject {subject} not found.')
    else:
        print('No data available.')
def view_top_students():
    if os.path.exists('studentdata.csv'):
        df = pd.read_csv('studentdata.csv')
        subject = input('Enter subject name: ')
        if subject in df['subject'].values:
            top_students = df[df['subject'] == subject].nlargest(3, 'marks')
            print(f'Top 3 students in {subject}:\n{top_students}')
        else:
            print(f'Subject {subject} not found.')
    else:
        print('No data available.')
def view_failing_students():
    if os.path.exists('studentdata.csv'):
        df = pd.read_csv('studentdata.csv')
        failing_students = df[df['marks'] < 40]
        if not failing_students.empty:
            print('Failing students:\n', failing_students)
        else:
            print('No failing students found.')
    else:
        print('No data available.')
def plot_marks_by_subject():
    if os.path.exists('studentdata.csv'):
        df = pd.read_csv('studentdata.csv')
        subject = input('Enter subject name: ')
        if subject in df['subject'].values:
            subject_data = df[df['subject'] == subject]
            plt.figure(figsize=(4,5))
            plt.bar(subject_data['name'], subject_data['marks'], color='blue')
            plt.xlabel('Students')
            plt.ylabel('Marks')
            plt.title(f'Marks Distribution in {subject}', fontsize=16, fontweight='bold', color='darkblue')
            plt.xticks(rotation=45)
            plt.show()
        else:
            print(f'Subject {subject} not found.')
    else:
        print('No data available.')
while True:
    menu()
    choice = int(input('Enter your choice(1-7): '))
    if choice == 1:
        add_result()
    elif choice == 2:
        view_data()
    elif choice == 3:
        view_subect_average()
    elif choice == 4:
        view_top_students()
    elif choice == 5:
        view_failing_students()
    elif choice == 6:
        plot_marks_by_subject()
    elif choice == 7:
        print('Exiting the program. Goodbye!')
        break
    else:
        print('Invalid choice, please try again.')
```
